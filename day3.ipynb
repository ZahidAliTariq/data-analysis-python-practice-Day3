```python
# for pandas library import
import pandas as pd
```

```python
# for read_csv data
df = pd.read_csv("student_performance_data_70.csv")
```

```python
# for check read_csv data
df
```

```python
# 9th class
df_class9 = df[df["Class"] == "9th"]
average_9th = df_class9["Math"] + df_class9["Science"] + df_class9["English"].mean()

print("Average of 9th class:")
print(df_class9)
```

```python
# 10th class
df_class10 = df[df["Class"] == "10th"]
average_10th = df_class10["Math"] + df_class10["Science"] + df_class10["English"].mean()

print("Average of 10th class:")
print(df_class10)
```

```python
# passed in lahore
df_pass_lahore = df[(df["City"] == "Lahore") & (df["Result"] == "Pass")]
print("Passed in Lahore:")
print(df_pass_lahore)
```

```python
# fail students
df_fail_students = df[df["Result"] == "Fail"]
print("Fail students are:")
print(df_fail_students)
```

```python
# student list Karachi
df_student_karachi = df[df["City"] == "Karachi"]
print("Student list of Karachi is:")
print(df_student_karachi)
```

```python
# top 10 students
df_top_students = df[df["Result"] == "Pass"].nlargest(10, ["Math", "Science", "English"])
print("Top 10 Students is:")
print(df_top_students)
```
